```python
import requests
import json
from github import Github
```

```python
def ruta(*args):
    url = 'https://api.github.com/'
    return url + '/'.join(args)
```

```python
url = ruta('users','ironhack-datalabs','repos')
response = requests.get(url)
repos_json = json.loads(response.text)
print(json.dumps(repos_json[0], indent=4))
```

```python
def get_key(key):
    with open('C:/Windows/AppsKeys/Github/'+key+'.txt', 'r') as file_read:
        app_key = file_read.readline()
    return app_key
```

```python
token = get_key('token')
git = Github(token)
git_user = git.get_user()
repo = git.get_repo('ironhack-datalabs/scavenger')
content = repo.get_content("")
```

```python
lista_paths = [carpeta.path for carpeta in content if carpeta.path != '.gitignore']
```

```python
lista_paths[:3]
```

```python
lista_archivos = []
for ruta in lista_paths:
    contenido_adentro = repo.get_contents(ruta)
    for archivo in contenido_adentro:
        if archivo.path.endswith('scavengerhunt'):
            lista_archivos.append(archivo.path)
contenido_adentro            
```

```python
rutas_ordenadas = sorted(lista_archivos, key= lambda x: x[9:11])
rutas_ordenadas
```

```python
for bueno in rutas_ordenadas:
    archivo = repo.get_contents(bueno)
    print(archivo.decoded_content)
```
